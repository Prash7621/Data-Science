```python
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
X, _ = make_blobs(n_samples=300, centers=4, n_features=4, random_state=42)
columns = ['Sales Amount', 'Number of Purchases', 'Category', 'Discount Rate']
sales_data = pd.DataFrame(X, columns=columns)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(sales_data)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_features)
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
plt.figure(figsize=(3, 2))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='blue', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Sales Dataset with Additional Feature')
plt.grid()
plt.show()
print(pca_df.head())
```
