```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy  import dendrogram,linkage
x=[4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y=[21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data=list(zip(x,y))

linkage_data=linkage(data,method='ward',metric='euclidean')
dendrogram(linkage_data)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

data=list(zip(x,y))
hierarchy_cluster=AgglomerativeClustering(n_clusters=2 ,metric='euclidean',linkage='ward')
labels=hierarchy_cluster.fit_predict(data)

plt.scatter(x,y,c=labels)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dt=pd.read_csv(r"D:\iris-data.csv")
dt.head(4)
```

```python
dt = dt.drop('class',axis='columns')
```

```python
sns.pairplot(data=dt)
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(dt,method='single',metric='euclidean'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=2,linkage='single')
dt['predict']=ac.fit_predict(dt)
dt.head(4)
```

```python
sns.pairplot(data=dt,hue="predict")
```
