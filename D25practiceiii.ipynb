```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

ds=pd.read_csv(r"D:\placement.csv")
ds.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa',y='package',data=ds)
```

```python
ds.isnull().sum()
```

```python
x=ds[["cgpa"]]
y=ds["package"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
Lr=LinearRegression()
Lr.fit(x_train,y_train)
```

```python
Lr.coef_,Lr.intercept_,Lr.predict([[5.12]]),Lr.score(x_train,y_train),Lr.score(x_test,y_test)
```

```python
y_pred=Lr.predict(x)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=ds)
plt.plot(ds['cgpa'],y_pred,c="red")
plt.legend(["org","predict"])
```
