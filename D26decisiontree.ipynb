```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt

data={
    'Age': [25,45,35,50,30],
    'Income':['low',"High",'Medium','Medium','Low'],
    'CreditScore' : ['Fair','Good','Good','Fair','Good'],
    'LoanApproved':['No','Yes','Yes','No','No']
    }

df=pd.DataFrame(data)

df['Income']=df['Income'].map({'Low':0,'Medium':1,'High':2})
df['CreditScore']=df['CreditScore'].map({'Fair':0,'Good':1})
df['LoanApproved']=df['LoanApproved'].map({'No':0,'Yes':1})


X=df[['Age','Income','CreditScore']]
y=df['LoanApproved']

model=DecisionTreeClassifier(criterion='gini', max_depth=3,random_state=42)
model.fit(X,y)


plt.figure(figsize=(10,6))
plot_tree(model,feature_names=['Age','Income','CreditScore'],class_names=['No','Yes'],filled=True,rounded=True)
plt.show()

new_applicant=[[60,1,1]]
prediction=model.predict(new_applicant)
print("LoanApproved?","Yes" if prediction[0]==1 else "No")
```
